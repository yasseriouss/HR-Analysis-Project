# Data **Distribution**

```python
numeric_cols = [
    'Age', 'MonthlyIncome', 'DistanceFromHome', 'YearsAtCompany',
    'TotalWorkingYears', 'YearsSinceLastPromotion', 'NumCompaniesWorked'
]
```

```python
fig, axes = plt.subplots(1,7, figsize=(22, 9))
fig.suptitle('Distribution Analysis — Histograms', fontsize=18, fontweight='bold')
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    axes[i].hist(df[col], bins=30, color='steelblue', edgecolor='white', alpha=0.85)
    axes[i].set_title(col, fontsize=12, fontweight='bold')
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Count')

    skew = df[col].skew()
    axes[i].annotate(
        f'Skew: {skew:.2f}',
        xy=(0.65, 0.88), xycoords='axes fraction',
        fontsize=9,
        color='red' if abs(skew) > 1 else 'green'
    )

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 7, figsize=(22, 9))
fig.suptitle('Outlier Detection — Box Plots', fontsize=18, fontweight='bold')
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    axes[i].boxplot(
        df[col], patch_artist=True,
        boxprops=dict(facecolor='steelblue', alpha=0.7),
        medianprops=dict(color='black', linewidth=2),
        flierprops=dict(marker='o', markerfacecolor='red', markersize=4, alpha=0.5)
    )
    axes[i].set_title(col, fontsize=12, fontweight='bold')
    axes[i].set_ylabel('Value')

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)).sum()

    axes[i].annotate(
        f'Outliers: {outliers}',
        xy=(0.5, 0.93), xycoords='axes fraction',
        fontsize=9, ha='center',
        color='red' if outliers > 20 else 'green'
    )

plt.tight_layout()
plt.show()
```

# Correlation

```python
df['Attrition_Binary'] = (df['Attrition'] == 'Yes').astype(int)

corr_cols = [
    'Age', 'MonthlyIncome', 'DistanceFromHome', 'YearsAtCompany',
    'TotalWorkingYears', 'YearsSinceLastPromotion', 'NumCompaniesWorked',
    'DailyRate', 'HourlyRate', 'MonthlyRate',
    'JobSatisfaction', 'EnvironmentSatisfaction',
    'WorkLifeBalance', 'JobInvolvement', 'JobLevel', 'Attrition_Binary'
]

corr = df[corr_cols].corr()

attrition_corr = corr['Attrition_Binary'].drop('Attrition_Binary')
high_corr_cols = attrition_corr[abs(attrition_corr) > 0.1].index.tolist()
high_corr_cols = high_corr_cols + ['Attrition_Binary']

print("الكولومنز اللي عندها علاقة قوية مع الـ Attrition:")
print(attrition_corr[abs(attrition_corr) > 0.1].sort_values(key=abs, ascending=False))
```

# T-**Test**

```python
from scipy.stats import ttest_ind, chi2_contingency, f_oneway

df['Attrition_Binary'] = (df['Attrition'] == 'Yes').astype(int)

# فصل الموظفين اللي اتركوا واللي فضلوا
stayed = df[df['Attrition'] == 'No']
left   = df[df['Attrition'] == 'Yes']

print("=" * 55)
print("  1. T-TEST — Numeric Columns")
print("=" * 55)

num_cols = [
    'Age', 'MonthlyIncome', 'DistanceFromHome',
    'YearsAtCompany', 'TotalWorkingYears',
    'JobSatisfaction', 'WorkLifeBalance',
    'EnvironmentSatisfaction', 'NumCompaniesWorked'
]

for col in num_cols:
    t_stat, p = ttest_ind(stayed[col], left[col])
    sig = "✅ Significant" if p < 0.05 else "❌ Not Significant"
    print(f"{col:<28} t={t_stat:+.3f}  p={p:.4f}  {sig}")
```

# Chi-**Square Test**

```python
print("\n" + "=" * 55)
print("  2. CHI-SQUARE — Categorical Columns")
print("=" * 55)

cat_cols = [
    'OverTime', 'MaritalStatus', 'BusinessTravel',
    'Gender', 'Department', 'JobRole', 'EducationField'
]

for col in cat_cols:
    ct = pd.crosstab(df[col], df['Attrition'])
    chi2, p, dof, _ = chi2_contingency(ct)
    sig = "✅ Significant" if p < 0.05 else "❌ Not Significant"
    print(f"{col:<28} chi2={chi2:.3f}  p={p:.4f}  {sig}")
```

# ANOVA **Test**

```python
print("\n" + "=" * 55)
print("  3. ANOVA — MonthlyIncome by Groups")
print("=" * 55)

anova_cols = ['Department', 'JobRole', 'MaritalStatus',
              'BusinessTravel', 'EducationField']

for col in anova_cols:
    groups = [df[df[col] == cat]['MonthlyIncome'].values
              for cat in df[col].unique()]
    f_stat, p = f_oneway(*groups)
    sig = "✅ Significant" if p < 0.05 else "❌ Not Significant"
    print(f"MonthlyIncome by {col:<18} F={f_stat:.3f}  p={p:.4f}  {sig}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import (classification_report, confusion_matrix,
                             roc_auc_score, roc_curve, ConfusionMatrixDisplay)
import matplotlib.pyplot as plt
import numpy as np

# ─── 1. جهّز الداتا ───────────────────────────────────────────
df_train = pd.read_csv('2-HR_Analytics (38 Columns).csv')

df_train['Attrition_Binary'] = (df_train['Attrition'] == 'Yes').astype(int)
df_train['OverTime_enc'] = (df_train['OverTime'] == 'Yes').astype(int)
df_train.drop_duplicates(inplace=True)
# الكولومنز الـ Significant من التحليل اللي فات
X1 = df_train[[
    'Age',
    'OverTime_enc',
    'DistanceFromHome',
    'JobSatisfaction',
    'WorkLifeBalance',
    'EnvironmentSatisfaction',
    'NumCompaniesWorked'
]]

y1 = df_train['Attrition_Binary']

# ─── 2. Train / Test Split ────────────────────────────────────
X_train, X_test, y_train, y_test = train_test_split(
    X1, y1, test_size=0.2, random_state=42, stratify=y
)

print(f"Train size : {len(X_train)} rows")
print(f"Test size  : {len(X_test)} rows")

# ─── 3. Scale ─────────────────────────────────────────────────
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc  = scaler.transform(X_test)
```

```python
# ─── 4. Logistic Regression ───────────────────────────────────
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train_sc, y_train)

print("\n── Logistic Regression ──────────────────────────────────")
print(f"AUC  : {roc_auc_score(y_test, lr.predict_proba(X_test_sc)[:,1]):.3f}")
print(classification_report(y_test, lr.predict(X_test_sc),
                             target_names=['Stay', 'Leave']))

# ─── 5. Random Forest ─────────────────────────────────────────
rf = RandomForestClassifier(n_estimators=200, random_state=42,
                             class_weight='balanced')
rf.fit(X_train, y_train)

print("── Random Forest ────────────────────────────────────────")
print(f"AUC  : {roc_auc_score(y_test, rf.predict_proba(X_test)[:,1]):.3f}")
print(classification_report(y_test, rf.predict(X_test),
                             target_names=['Stay', 'Leave']))
```
